```python
import numpy as np
import matplotlib.pyplot as plt

def drawdown(p):
    cum_rets = p.cumsum()+1

    # Calculate the running maximum
    running_max = np.maximum.accumulate(cum_rets.dropna())

    # Ensure the value never drops below 1
    running_max[running_max < 1] = 1

    # Calculate the percentage drawdown
    drawdown = (cum_rets)/running_max - 1
    drawdown = np.where(drawdown<-1, -1, drawdown)
    return -np.min(drawdown)

drawdown_ = []
sharpe = []
idx = []
for col in f.columns:
  try:
    drawdown_.append(drawdown(f[col].dropna()))
    sharpe.append(np.sqrt(252)*f[col].dropna().mean()/f[col].dropna().std())
    idx.append(col)
  except:
    pass

values = pds.DataFrame([drawdown_,sharpe], index=["Drawdown", "Sharpe"], columns=idx).transpose()

def distance(x,y):
  if np.shape(x)!=np.shape(y):
    print(ValueError("We need array of the same shape"))
  else:
    return np.sqrt(np.sum(np.abs(x.transpose()-y.transpose()), axis=0))

distance(values.iloc[0:1].values,values.iloc[5:7].values)

b = np.array([[0.07,1.35]])
dis = []
for i in range(len(values)):
  dis.append(distance(values.iloc[i:i+1,:].values, b)[0])

r = pds.DataFrame(dis,columns=["Point"], index=values.index).sort_values(by="Point", ascending=True).iloc[0:10,:].index
values["label"] = 0
values.loc[r,"label"]=1
```

```python
f = yf.download(actifs, "2015-01-01", "2021-01-01")["Adj Close"].pct_change(1).dropna()
def SR_criterion(weight, returns, rf):
  """ 
  ------------------------------------------------------------------------------
  | Output: Opposite Sharpe ratio to do a m imization                          |
  ------------------------------------------------------------------------------
  | Inputs: -Weight (type ndarray numpy): Wheight for portfolio                |
  |         -returns (type dataframe pandas): Returns of stocks                |
  |         -rf (float): risk-free asset                                       |
  ------------------------------------------------------------------------------
  """
  pf_return = returns.dot(weight)
  mu = np.mean(pf_return) - rf
  sigma = np.std(pf_return)
  Sharpe = -mu/sigma
  return Sharpe

def max_sharpe_optimisation(returns):
  """
  Optimisation
  """
  short_selling_allowed = True
  from scipy.optimize import minimize
  n = len(returns.transpose())
  # Initialisation weight value
  x0 = np.zeros(n)+(1/n)

  # Optimization constraints problem
  cons=({'type':'eq', 'fun': lambda x:sum(abs(x))-1})

  if short_selling_allowed == True:
      Bounds= [(-1 , 1) for i in range(0,n)]
  else:
      Bounds= [(0 , 1) for i in range(0,n)]

  # Lambda
  Lambda_RA=3

  # Optimization problem solving
  res_SR = minimize(SR_criterion, x0, method="SLSQP", args=(returns, 0),bounds=Bounds,constraints=cons,options={'disp': False})

  # Result for visualization
  #print(pds.DataFrame(np.round(res_SR.x,5),index=returns.columns, columns=["Weihgt"]))

  # Result for computations
  X = res_SR.x

  sharpe = SR_criterion(X, returns, 0)*np.sqrt(252)
  return X

class features():

  def __init__(self, var, lag):

    self.var = var
    self.col = var.columns
    self.arr = var.values
    self.lag = lag
  
  def add_covariances(self):
    col=[]
    covariances = []
    for i in range(self.lag, len(f)):
        self.cov = []
        
        sub_arr = np.cov(self.arr[i-self.lag:i,:],rowvar=False)
        for r in range(len(sub_arr)):
          for c in range(len(sub_arr)):
            if r>=c:
              self.cov.append(sub_arr[r][c])
            else:
              pass
        covariances.append(self.cov)

    for r in range(len(sub_arr)):
          for c in range(len(sub_arr)):
            if r>=c:
              if r!=c:
                col.append(f"Cov {f.columns[r]}/{f.columns[c]}")
              else:
                col.append(f"Var {f.columns[r]}")
    covariances =  pds.DataFrame(covariances, columns=col, index=self.var.index[self.lag:])
    self.var = pds.concat((self.var, covariances), axis=1)


  def add_weights(self):
    l = []
    for i in range(len(self.var)-self.lag):
      h = self.arr[i:i+self.lag,:]
      p = max_sharpe_optimisation(h)
      l.append(p)
    s = pds.DataFrame(l, index=self.var.index[:-self.lag], columns=[f"Weight {col} " for col in self.col])
    self.var = pds.concat((self.var, s), axis=1)


  def add_percentage(self):
    for col in self.col:
      self.var[f"{col} pct+lag"] = self.var[col].rolling(self.lag).sum().shift(-self.lag)
      print(self.col)
  
  def add_weights_p(self):
    for col in self.col:
      self.var[f"P weights {col}"] = self.var[f"Weight {col} "].shift(self.lag)

  def add_mean(self):
    mea = list()
    for i in range(self.lag, len(f)):
        sub_arr = np.mean(self.arr[i-self.lag:i,:], axis=0)
        mea.append(sub_arr)
    means = pds.DataFrame(mea, columns=[f"mean {col}" for col in self.col], index=self.var.index[self.lag:])
    self.var = pds.concat((self.var,means), axis=1)

  
F = features(f[actifs],20)
F.add_covariances()
F.add_weights()
F.add_percentage()
F.add_weights_p()
F.add_mean()

F.var
```

```python
f = F.var
def explained_variance(f):
    """
    +----------------------------------------------------------------------------------------------------------+
    |Description:  Renvoie un graphique reliant la variance expliquée et le nombre de variable utilisée (ACP)  |
    +----------------------------------------------------------------------------------------------------------+
    |Paramètre: - f: base de données                                                                           |
    |                                                                                                          |
    +----------------------------------------------------------------------------------------------------------+
    |Sortie: Aucune variable mais un graphique affichant la relation en variance expliqué et nombre de variable|       |
    +----------------------------------------------------------------------------------------------------------+
    """
    import numpy as np
    from sklearn.decomposition import PCA
    import plotly.graph_objs as go

    df = f
    pca = PCA()
    pca.fit(df)
    exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)
    O = [0 for i in range(len(exp_var_cumul)+1)]
    U = [i for i in range(len(exp_var_cumul))]
    
    y1_upper = [0]+[v for v in exp_var_cumul]
    y1_lower = O
    y1_lower = y1_lower[::-1]

    x = [i for i in range(len(U)+1)]
    x_rev = x[::-1]
    fig = go.Figure()

    fig.add_trace(go.Scatter(
    x=x+x_rev,
    y=np.concatenate((y1_upper,y1_lower),axis=0),
    fill='toself',
    fillcolor='rgba(0,100,80,0.2)',
    line_color='rgba(255,255,255,0)',
    showlegend=False,
    name="confiance",
    mode='lines'
    ))

    fig.add_trace(go.Scatter(
      x=U, y=y1_upper,
            mode="markers",
                  name="% of Information",
                        line=dict(color="#53A7B9",width=15, dash='dot'),
                        marker=dict(size=10,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
     
    ))
    fig.show()

explained_variance(f.dropna())
```

```python

def drawdown(p):
    cum_rets = p.cumsum()+1

    # Calculate the running maximum
    running_max = np.maximum.accumulate(cum_rets.dropna())

    # Ensure the value never drops below 1
    running_max[running_max < 1] = 1

    # Calculate the percentage drawdown
    drawdown = (cum_rets)/running_max - 1
    drawdown = np.where(drawdown<-1, -1, drawdown)
    return -np.min(drawdown)

def sortino(p):
  p_b = p.loc[p[p.columns[0]]<0]
  sigma = np.std(p_b)
  return (np.sqrt(12)*p.mean()/sigma).values[0]

def var(p):
  p = p.sort_values(by=p.columns[0], ascending=False)
  split = int(len(p)*0.01)
  return np.max(p.iloc[0:split])[0]
min_drawdown = []
sharpe = []
Var = []
mean = []
volat = []
Sortino = []
for col in alg:
  min_drawdown.append(drawdown(g[col]/F.lag))
  mean.append(12*(g[col]).mean())
  volat.append(np.sqrt(12)*(g[col]).std())
  sharpe.append(np.sqrt(12)*(g[col]).mean()/(g[col]).std())
  Sortino.append(sortino(g[[col]]))
  Var.append(var(g[[col]]))

pds.DataFrame([min_drawdown, mean, volat, sharpe, Sortino, Var], index=["Min Darwdown", "Mean", "Volatility", "Sharpe", "Sortino", "Var"], columns=alg)
```
